```python
import os

import requests
from tabula import read_pdf
```

# Read data from PDF tables

**Note:** This notebook does not perform OCR. It reads structured text (not images) from a PDF.

```python
# The URL of the PDF file to download
url = "https://niva.brage.unit.no/niva-xmlui/bitstream/handle/11250/3126040/7963-2024.pdf?sequence=1&isAllowed=y"

# File to create
filename = "M2730.pdf"

# Page to read
page_no = 14
```

```python
# Download temporary copy of PDF
try:
    response = requests.get(url, stream=True)
    response.raise_for_status()

    with open(filename, "wb") as file:
        for chunk in response.iter_content(chunk_size=8192):
            file.write(chunk)

    print(f"Successfully downloaded the PDF to '{filename}'")

except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```

```python
# Read a table from the PDF
df = read_pdf(filename, pages=page_no)[0]

# Remove local copy
os.remove(filename)

df.head(30)
```
