# Query data via the Vann-Nett API

[Vannmiljø](https://vannmiljo.miljodirektoratet.no/) is Miljødirektoratet's national water quality database. [Vann-Nett](https://vann-nett-klient.miljodirektoratet.no/waterbodies/map) uses data from Vannmiljø to perform water quality status classifications e.g. for the Water Framework Directive. Vann-Nett does not store raw data, but rather presents aggregated "indexes" for biologic, hydromorphologic and physico-chemical parameters (including WFD class boundaries etc.)

Vann-Nett does not have an "official" public API, but inspection of the network traffic when the website loads shows that it does have a public API, even though it is not advertised by MDir. Relevant endpoints are:

 * `https://vann-nett.no/service/waterbodies/{wb_id}/qualityElements/ecological`. Ecological quality elements (including nutrients etc.).
 * `https://vann-nett.no/service/waterbodies/{wb_id}/qualityElements/RBSP`. River Basin Specific Pollutants (RBSP).
 * `https://vann-nett.no/service/waterbodies/{wb_id}/qualityElements/swChemical`. Chemical quality elements.

(There are probably others too - these are just the ones I could find quickly).

This notebook provides basic functionality for querying this API.

```python
wb_id = "003-65-R"
quality_element = "ecological"
nivapy.da.get_data_from_vannnett(wb_id, quality_element=quality_element)
```
