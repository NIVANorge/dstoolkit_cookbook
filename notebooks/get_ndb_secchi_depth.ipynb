```python
import nivapy3 as nivapy
import pandas as pd

# Connect to nivabasen
eng = nivapy.da.connect()

# Period of interest
st_dt = "1900-01-01"
end_dt = "2022-12-31"

# Get all stations
stn_df = nivapy.da.select_ndb_stations(eng)

# Get all params in period for stations
par_df = nivapy.da.select_ndb_station_parameters(stn_df, st_dt, end_dt, eng)

# Subset to params of interest
par_df = par_df.query("parameter_name == 'Siktdyp'")

# Get data
wc_df, dup_df = nivapy.da.select_ndb_water_chemistry(
    stn_df, par_df, st_dt, end_dt, eng, lod_flags=False, drop_dups=True
)

# Save
wc_df.to_csv("nivadatabase_secchi_depth.csv", index=False)
```
